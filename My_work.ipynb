```python
import datetime
```

```python
def month(x):
    month_name = x
    datetime_object = datetime.datetime.strptime(month_name, "%b")
    month_number = datetime_object.month
    return month_number
```

```python
df['Month']=df['Month'].apply(month)
```

## Machine Learning

Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
df.columns
```

```python
X = df[['How_many_Landed_on_our_Page?',
       'How_many_Landed_on_the_our_Page_and_clicked_on_a_button?',
       'How_many_Landed_on_the_our_Page_and_clicked_on_a_button_and_started_filling_the_Form?']]
```

```python
y = df['How_many_Landed_on_the_our_Page_and_clicked_on_a_button_and_started_filling_the_Form_and_Completed_and_submited_the_form?']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

Training the model

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

```python
pred = lm.predict(X_test)
```

```python
y_pred = pd.DataFrame(pred)
```

```python
plt.scatter(y_test, pred)
```

Evaluate the model

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
print('MAE', mean_absolute_error(y_test, pred))
```

```python
print('MSE', mean_squared_error(y_test, pred))
```

```python
print('RMSE', np.sqrt(mean_squared_error(y_test, pred)))
```


Resiuduals

```python
sns.displot((y_test-pred), bins = 40)
```
